```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
np.set_printoptions(precision=4)
```

```python
X = np.random.randn(2,50) + 1
X = X * [[1],[0.1]]
```

```python
fig, ax = plt.subplots()
ax.scatter(X[0],X[1])
ax.grid()
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.axis('equal')
plt.show()
```

```python
u = np.mean(X, axis=1, keepdims=True)
X = X - u

```

```python
fig, ax = plt.subplots()
ax.scatter(X[0],X[1])
ax.grid()
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.axis('equal')
plt.show()
```

```python
o = np.square(X)
o = np.sum(o, axis=1, keepdims=True)
o = o / X.shape[1]
#o = np.sqrt(o)
print(o)
print(tf.math.reduce_variance(X,axis=1))
X = X / o
```

```python
fig, ax = plt.subplots()
ax.scatter(X[0],X[1])
ax.grid()
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
plt.xlim(-5,5)
plt.ylim(-5,5)
ax.set_aspect('equal', adjustable='box')
plt.show()
```

```python
n = np.square(X)
n = np.sum(n, axis=1, keepdims=True)
n = np.sqrt(n)
X = X / n
```

```python
fig, ax = plt.subplots()
ax.scatter(X[0],X[1])
ax.grid()
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
plt.xlim(-1,1)
plt.ylim(-1,1)
ax.set_aspect('equal', adjustable='box')
plt.show()
```

```python
print(np.sqrt(np.sum(np.square(X), axis=1, keepdims=True)))
```
