```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.random.randn(2,50) + 1
X = X * [[1],[0.1]]
```

```python
fig, ax = plt.subplots()
ax.scatter(X[0],X[1])
ax.grid()
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.axis('equal')
plt.show()
```

```python
u = np.sum(X, axis=1, keepdims=True)
u = u / X.shape[1]
X = X - u

```

```python
fig, ax = plt.subplots()
ax.scatter(X[0],X[1])
ax.grid()
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.axis('equal')
plt.show()
```

```python
o = np.square(X)
o = np.sum(o, axis=1, keepdims=True)
o = np.sqrt(o)
o = o / X.shape[1]
X = X / o
```

```python
fig, ax = plt.subplots()
ax.scatter(X[0],X[1])
ax.grid()
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.axis('equal')
plt.show()
```
