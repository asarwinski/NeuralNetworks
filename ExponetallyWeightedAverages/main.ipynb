```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,20, 0.1)
y = np.sin(x) + 10
print(y.shape)
```

```python
plt.plot(y)
plt.show()
```

```python
noise = np.random.randn(y.shape[0])
y_noise = y + noise
```

```python
plt.plot(y_noise)
plt.show()
```

```python
y_ewa = []
temp = 0
beta = 0.9

for i in range(y_noise.shape[0]):
    temp = beta*temp + (1-beta)*y_noise[i]
    y_ewa.append(temp / (1 - beta**(i+1)))
```

```python
plt.plot(y_noise)
plt.plot(y_ewa)
#plt.plot(y_ewa2)
plt.show()
```
